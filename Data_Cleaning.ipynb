```python
Airbnb['price'] = pd.to_numeric(Airbnb['price'], errors='coerce')
sns.set_theme(rc={"figure.figsize": (20, 6)})
sns.histplot(Airbnb['price'], kde=False, bins=30)
```

```python
print("Skewness: %f" % Airbnb['price'].skew())
print("Kurtosis: %f" % Airbnb['price'].kurt())
```

As the price column has skewness has value >1, it is posetivly skewed.

 Also high value of kurosis also indicate it has good amount of outlier

```python
#using IQR method to detect and remove outliers
q1 = Airbnb['price'].quantile(0.25)
q3 = Airbnb['price'].quantile(0.75)
iqr = q3 - q1

lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

New_Air=Airbnb[(Airbnb['price'] >= lower_bound) & (Airbnb['price'] <= upper_bound)]
New_Air
```

```python

sns.histplot(New_Air['price'], kde=False, bins=30)
```

```python
# Finding Which Neighbour Group has Most costliest place to live in

sns.violinplot(x='neighbourhood_group',y='price',data=New_Air)
plt.title("Neighbourhood_group wise price distributions")
```

As we can see Manhattan is most costliest in Neighbourhood_group in the list. 

Brooklyn and Queens have almost same price distribution.

Staten Island does not vary much as its price is almost fix like in between (50-100)

Bronx price is the cheapest as its price is less than 50

```python
# Find which type of room is in demand

New_Air.groupby('room_type')['minimum_nights'].mean().plot(kind='barh',color='g')
plt.title('Min Stays in different room types listed on Airbnb ')
plt.xlabel('Minimum Night Stays')
plt.ylabel('Room Types')
```

From the above plot its clear that people like to stay in Entire home/apt compared to private and shared room

```python
# find Which type property has majority in which neighbour_group

plt.rcParams['figure.figsize'] = (8, 5)
ax= sns.countplot(y='room_type',hue='neighbourhood_group',data=New_Air,palette='bright')

total = len(New_Air['room_type'])
for p in ax.patches:
        percentage = '{:.1f}%'.format(100 * p.get_width()/total)
        x = p.get_x() + p.get_width() + 0.02
        y = p.get_y() + p.get_height()/2
        ax.annotate(percentage, (x, y))

plt.title('Count of each room types in entire Airbnb')
plt.xlabel('Percentage acquired in neighbour_group')
plt.xticks(rotation=90)
plt.ylabel('Room Types')

plt.show()
```

From Above plot it can be assumed that :

 1.Manhatten has most property of Entire Home/Apt

 2.Brooklyn is slightly ahead in case of property of Private room than Manhattan

 3.Shared room only available in Manhattan
 
 4.Queens has few property of Private room and Entire room/Apt

```python
HR=pd.read_csv(input_path+"\\HR_Dataset.csv")
HR.head()
```

```python
HR['DOB'] = pd.to_datetime(HR['DOB'], errors='coerce')
HR['DateofHire'] = pd.to_datetime(HR['DateofHire'], errors='coerce')
HR['DateofTermination'] = pd.to_datetime(HR['DateofTermination'], errors='coerce')
HR['LastPerformanceReview_Date'] = pd.to_datetime(HR['LastPerformanceReview_Date'], errors='coerce')

# Checking if ManagerID can now be converted to integer
HR['ManagerID'] = HR['ManagerID'].astype(int)

# Dropping irrelevant columns (e.g., EmpID, Zip)
HR_cleaned = HR.drop(columns=['EmpID', 'Zip', 'PositionID'])
```

```python
HR_cleaned.info(), 
```

```python
# Find the salary distribution of this industry

plt.figure(figsize=(10,6))
sns.kdeplot(HR_cleaned['Salary'], fill=True, color="g")
plt.title("Kernel Density Estimation of Salary Distribution")
plt.xlabel("Salary")
plt.show()
```

From this plot it shows that:
1. The maximum no employee working it this industry have the salary range from 50000-90000.
2. The salary range from 130000 -250000 have been distributed to very small amount of manpower of this industry.

```python
#Find the percantage of the total employee meets the expectation of the company
performance_counts = HR_cleaned['PerformanceScore'].value_counts()

plt.figure(figsize=(8,8))
plt.pie(performance_counts, labels=performance_counts.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('muted'))
plt.title("Performance Score Distribution")
plt.axis('equal')
plt.show()
```

From the Graph it showing:
1. 78.1 % of total manpower  fully meets and 11.9% of the total manpower exceeds the company's expectation
2. 5.8% of total manpower has the ability to improve.
3. for 4.2% of total manpower company perpared PIP ( Performance Improvment Plan) for them.

```python
# Find the hiring and termination trend of this company
HR_cleaned['YearofTermination'] = HR_cleaned['DateofTermination'].dt.year
hire_counts = HR_cleaned['YearofHire'].value_counts().sort_index()
term_counts = HR_cleaned['YearofTermination'].value_counts().sort_index()

plt.figure(figsize=(10,6))
plt.fill_between(hire_counts.index, hire_counts.values, color="skyblue", alpha=0.4, label="Hires")
plt.fill_between(term_counts.index, term_counts.values, color="orange", alpha=0.4, label="Terminations")
plt.title("Hiring and Termination Trends Over Time")
plt.xlabel("Year")
plt.ylabel("Number of Employees")
plt.legend(loc='upper left')
plt.grid(True)
plt.show()
```

From this graph:
1. This company has posetivly hired from 2006 to 2010 and no termination
2. After 2010 company hired a lot of  employee as well as terminated few employees
3. In betweem 2016 to 2018 company has terminated many emplyees and no new hiring

```python
# Find Percantage of absent wrt Department
absences_per_dept = HR_cleaned.groupby('Department')['Absences'].sum()

plt.figure(figsize=(8,8))
plt.pie(absences_per_dept, labels=absences_per_dept.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))


centre_circle = plt.Circle((0, 0), .50, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title("Employee Absences by Department")
plt.axis('equal')  
plt.show()
```

From the chart we can conclude:
1. Production Department has the maximum nos of day of absent
2. It/IS is in 2nd position for absent and sales department also having 11.2% of absent rate

```python
heatmap_data = HR_cleaned.pivot_table(values='DaysLateLast30', index='Department', aggfunc='sum')

# Create a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(heatmap_data, annot=True, fmt=".0f", cmap='Blues', cbar_kws={'label': 'Total Days Late'})
plt.title("Heatmap of Total Days Late by Department")
plt.xlabel("Total Days Late")
plt.ylabel("Department")
plt.show()
```

From the above chart:
1. Production Department makes most late entry in the last 30 days 
2. Sales Department comes in 2nd who makes late entry
